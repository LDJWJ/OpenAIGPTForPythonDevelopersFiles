```python
import autogen
import os

# OpenAI API í‚¤ë¥¼ í™˜ê²½ ë³€ìˆ˜ì—ì„œ ê°€ì ¸ì˜¤ê¸°
OPENAI_APIKEY = os.environ.get("API_KEY")  # API_KEY -> OPENAI_API_KEYë¡œ ë³€ê²½

class MarketingAgentSystem:
    def __init__(self):
        # OpenAI API í‚¤ ì„¤ì • (í™˜ê²½ ë³€ìˆ˜ ì‚¬ìš© ê¶Œì¥)
        self.config_list = [
            {
                "model": "gpt-4o-mini",
                "api_key": OPENAI_APIKEY
            }
        ]

        # LLM ì„¤ì •
        self.llm_config = {
            "config_list": self.config_list,
            "temperature": 0.7,
        }

        # ì—ì´ì „íŠ¸ ì´ˆê¸°í™”
        self.market_researcher = autogen.AssistantAgent(
            name="market_researcher",
            system_message="ë‹¹ì‹ ì€ ì „ë¬¸ ì‹œì¥ ì¡°ì‚¬ ì „ë¬¸ê°€ì…ë‹ˆë‹¤. ì œí’ˆì— ëŒ€í•œ ìƒì„¸í•œ ì‹œì¥ ì¡°ì‚¬ ë³´ê³ ì„œë¥¼ ì‘ì„±í•˜ì„¸ìš”.",
            llm_config=self.llm_config
        )

        self.customer_segmentation = autogen.AssistantAgent(
            name="customer_segmentation_expert",
            system_message="ë‹¹ì‹ ì€ ê³ ê° ì„¸ë¶„í™” ì „ë¬¸ê°€ì…ë‹ˆë‹¤. ì‹œì¥ ì¡°ì‚¬ ê²°ê³¼ë¥¼ ë°”íƒ•ìœ¼ë¡œ ê³ ê°ì¸µì„ ì„¸ë¶„í™”í•˜ê³  ë¶„ì„í•˜ì„¸ìš”.",
            llm_config=self.llm_config
        )

        self.marketing_strategist = autogen.AssistantAgent(
            name="marketing_strategist",
            system_message="ë‹¹ì‹ ì€ ìµœê³ ì˜ ë§ˆì¼€íŒ… ì „ëµ ì „ë¬¸ê°€ì…ë‹ˆë‹¤. ì‹œì¥ ì¡°ì‚¬ì™€ ê³ ê° ì„¸ë¶„í™” ê²°ê³¼ë¥¼ ì¢…í•©í•˜ì—¬ ë§ˆì¼€íŒ… ì „ëµì„ ìˆ˜ë¦½í•˜ì„¸ìš”.",
            llm_config=self.llm_config
        )

        self.user_proxy = autogen.UserProxyAgent(
            name="user_proxy",
            human_input_mode="NEVER",
            max_consecutive_auto_reply=10
        )

    def generate_marketing_report(self, product_concept):
        # ì‹œì¥ ì¡°ì‚¬ ë‹¨ê³„
        self.user_proxy.initiate_chat(
            self.market_researcher,
            message=f"{product_concept}ì— ëŒ€í•œ ì‹œì¥ ì¡°ì‚¬ ë³´ê³ ì„œë¥¼ ì‘ì„±í•´ì£¼ì„¸ìš”.",
            max_turns=1
        )
        # last_message()ì—ì„œ 'content' í‚¤ë¡œ ì ‘ê·¼
        market_research_result = self.market_researcher.last_message()['content']

        # ê³ ê° ì„¸ë¶„í™” ë‹¨ê³„
        self.user_proxy.initiate_chat(
            self.customer_segmentation,
            message=f"ë‹¤ìŒ ì‹œì¥ ì¡°ì‚¬ ê²°ê³¼ë¥¼ ë°”íƒ•ìœ¼ë¡œ ê³ ê° ì„¸ë¶„í™” ë³´ê³ ì„œë¥¼ ì‘ì„±í•´ì£¼ì„¸ìš”: {market_research_result}",
            max_turns=1
        )
        # last_message()ì—ì„œ 'content' í‚¤ë¡œ ì ‘ê·¼
        customer_segmentation_result = self.customer_segmentation.last_message()['content']

        # ë§ˆì¼€íŒ… ì „ëµ ìˆ˜ë¦½ ë‹¨ê³„
        self.user_proxy.initiate_chat(
            self.marketing_strategist,
            message=f"""
            ë‹¤ìŒ ì •ë³´ë¥¼ ë°”íƒ•ìœ¼ë¡œ ì¢…í•©ì ì¸ ë§ˆì¼€íŒ… ì „ëµì„ ìˆ˜ë¦½í•´ì£¼ì„¸ìš”:
            ì‹œì¥ ì¡°ì‚¬: {market_research_result}
            ê³ ê° ì„¸ë¶„í™”: {customer_segmentation_result}
            """,
            max_turns=1
        )
        # last_message()ì—ì„œ 'content' í‚¤ë¡œ ì ‘ê·¼
        marketing_strategy_result = self.marketing_strategist.last_message()['content']

        return {
            "product_concept": product_concept,
            "market_research": market_research_result,
            "customer_segmentation": customer_segmentation_result,
            "marketing_strategy": marketing_strategy_result
        }

def main():
    product_concept = "AI ê¸°ë°˜ ê°œì¸ ë§ì¶¤í˜• ê±´ê°• ê´€ë¦¬ ì›¨ì–´ëŸ¬ë¸” ë””ë°”ì´ìŠ¤"

    marketing_system = MarketingAgentSystem()
    marketing_report = marketing_system.generate_marketing_report(product_concept)

    print("ğŸš€ ë¯¸ë˜ ìœ ë§ ì•„ì´í…œ ë§ˆì¼€íŒ… ì „ëµ ë³´ê³ ì„œ ğŸš€")
    print("\n1. ì œí’ˆ ê°œë…:")
    print(marketing_report['product_concept'])

    print("\n2. ì‹œì¥ ì¡°ì‚¬ ê²°ê³¼:")
    print(marketing_report['market_research'])

    print("\n3. ê³ ê° ì„¸ë¶„í™”:")
    print(marketing_report['customer_segmentation'])

    print("\n4. ë§ˆì¼€íŒ… ì „ëµ:")
    print(marketing_report['marketing_strategy'])

if __name__ == "__main__":
    main()
```
